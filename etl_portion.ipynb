```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
from census import Census

import numpy as np

import requests
```

```python
# Census & gmaps API Keys
from config import (api_key, gkey)
c = Census(api_key, year=2013)
```

```python
# File Path
covid_data_file = "FlaskApp/resource/COVID-19_Cases__Tests__and_Deaths_by_ZIP_Code.csv"

# Read CSV
covid_data = pd.read_csv(covid_data_file, thousands=',')

# Show Data
covid_data
```

```python
covid_data.index.name ='ID'
covid_data
```

```python
covid_data = covid_data.sort_values(by=['ZIP Code', 'Week Number'])
covid_data.head()
```

```python
covid_data=covid_data.drop(["Case Rate - Weekly", "Test Rate - Weekly", "Test Rate - Cumulative","Death Rate - Weekly","Death Rate - Cumulative"], axis=1)
covid_data
```

```python
covid_data.reset_index(inplace=True)
covid_data
```

```python
covid_data = covid_data.fillna(0)
covid_data
```

```python
covid_data["Positive Test Rate"] = (100 * \
    covid_data["Cases - Cumulative"].astype(
        int) / covid_data["Tests - Cumulative"].astype(int)).round(2)
covid_data.head()
```

```python
covid_data["Death Rate"] = (100 * \
    covid_data["Deaths - Cumulative"].astype(
        int) / covid_data["Cases - Cumulative"].astype(int)).round(2)
covid_data.head()
```

```python
covid_data['Week Start']=covid_data['Week Start'].astype(str)
covid_data
```

```python
covid_data = covid_data.fillna(0)
covid_data
```

```python
covid_data=covid_data.drop("ID", axis=1)
covid_data
```

```python
covid_data.index.name ='ID'
covid_data
```

```python
# Load poverty and race info from census API
census_race_data = c.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E",
                          "B19301_001E",
                          "B17001_002E","B01003_001E", "B02001_002E", "B02001_003E", "B02001_004E", "B02001_005E", "B02001_006E", "B02001_007E", "B02001_008E"), {'for': 'zip code tabulation area:*'})
census_data = pd.DataFrame(census_race_data)
census_data.head()
```

```python
# Rename the columns
census_data = census_data.rename(columns={"B01003_001E": "Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income",
                                      "B17001_002E": "Poverty Count",
                                      "B02001_002E": "White",
                                      "B02001_003E": "Black or African American",
                                      "B02001_004E": "American Indian and Alaska Native",
                                      "B02001_005E": "Asian",
                                      "B02001_006E": "Hawaiian and Other Pacific Islander",
                                      "B02001_007E": "Other",
                                      "B02001_008E": "Two or More Mixed",
                                      "NAME": "Name", "zip code tabulation area": "Zipcode"})
census_data.head()
```

```python
census_data = census_data.rename(columns={"Zipcode": "ZIP Code"})
census_data
```

```python

# Add in Poverty Rate (Poverty Count / Population)
census_data["Poverty Rate"] = 100 * \
    census_data["Poverty Count"].astype(
        int) / census_data["Population"].astype(int)
census_data
```

```python
# Final DataFrame
census_data = census_data[["ZIP Code", "Population", "Median Age", "Household Income",
                       "Per Capita Income", "Poverty Count", "Poverty Rate", "White", "Black or African American","Asian", "American Indian and Alaska Native", "Hawaiian and Other Pacific Islander","Two or More Mixed", "Other"]]
census_data.head()
```

```python
census_data['ZIP Code']=census_data['ZIP Code'].astype(int)
```

```python
zipcodes_list=covid_data['ZIP Code']
zipcodes_list=pd.DataFrame(zipcodes_list)
zipcodes_list
```

```python
zipcodes_list['ZIP Code']=pd.to_numeric(zipcodes_list['ZIP Code'], errors='coerce')
```

```python
# Merge zipcode and census data
census_zipcode_merge_df = census_data.merge(zipcodes_list, how='inner', left_on=["ZIP Code"], right_on=["ZIP Code"] )
census_zipcode_merge_df
```

```python
census_zipcode_merge_df.index.name ='ID'
census_zipcode_merge_df
```

```python
# Set Connection String
connection_string = "postgres:postgres@localhost:5432/covid_db"

# Set Engine
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm Tables
engine.table_names()
```

```python
# Fill Covid Table
covid_data.to_sql(name='covid', con=engine, if_exists='append', index=True)
```

```python
# Fill Census
census_zipcode_merge_df.to_sql(name='census', con=engine, if_exists='append', index=True)
```
