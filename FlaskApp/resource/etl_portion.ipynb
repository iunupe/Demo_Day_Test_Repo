```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
from census import Census

import numpy as np

import requests
```

```python
# Census & gmaps API Keys
from config import (api_key, gkey)
c = Census(api_key, year=2013)
```

```python
# File Path
covid_data_file = "COVID-19_Cases__Tests__and_Deaths_by_ZIP_Code.csv"

# Read CSV
covid_data = pd.read_csv(covid_data_file)

# Filter Out Years After 2015
#unemployment_df = unemployment_df[unemployment_df["Year"] <= 2015]

# Show Data
covid_data
```

```python
covid_data.index.name ='ID'
covid_data
```

```python
# File Path
census_data_file = "census_data.csv"

# Read CSV
census_data = pd.read_csv(census_data_file)

# Show Data
census_data.head()
```

```python
census_data = census_data.rename(columns={"Zipcode": "ZIP Code"})
census_data
```

```python
census_data['ZIP Code']=census_data['ZIP Code'].astype(int)
```

```python
zipcodes_list=covid_data['ZIP Code']
zipcodes_list=pd.DataFrame(zipcodes_list)
zipcodes_list
```

```python
zipcodes_list['ZIP Code']=pd.to_numeric(zipcodes_list['ZIP Code'], errors='coerce')
```

```python
# Merge zipcode and census data
census_zipcode_merge_df = census_data.merge(zipcodes_list, how='inner', left_on=["ZIP Code"], right_on=["ZIP Code"] )
census_zipcode_merge_df
```

```python
census_zipcode_merge_df.index.name ='ID'
census_zipcode_merge_df
```

```python
# Set Connection String
connection_string = "postgres:postgres@localhost:5432/covid_db"

# Set Engine
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm Tables
engine.table_names()
```

```python
# Fill Covid Table
covid_data.to_sql(name='covid', con=engine, if_exists='append', index=True)
```

```python
# Fill Census
census_zipcode_merge_df.to_sql(name='census', con=engine, if_exists='append', index=True)
```
